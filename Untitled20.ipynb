```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
from transformers import pipeline
import gradio as gr
```

```python


pipe = pipeline("text-generation", model="Qwen/Qwen3-0.6B")
# messages = [
#     {"role": "user", "content": "Who are you?"},
# ]
# pipe(messages)
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={"max_new_tokens": 2048, "max_length": 4096})
```

```python
chat_model=ChatHuggingFace(llm=llm)
```

```python
messages=[
    {"role":"user","content":"Who are you?"}
]
```

```python
response=chat_model.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_template("Answer as a scientist: {question}")
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
chain=prompt | chat_model|StrOutputParser()
```

```python
full_response=chain.invoke({"question":"What is Deep Learning?"})
```

```python
final_answer = full_response.split("</think>")[-1].strip() if "</think>" in full_response else full_response.strip()
print(final_answer)
```

# gradio

```python
def generate_answer(question):
    messages = [("user", question)]
    full_response = chain.invoke(messages)

    final_answer = full_response.split("</think>")[-1].strip() if "</think>" in full_response else full_response.strip()
    return final_answer

demo = gr.Interface(
    fn=generate_answer,
    inputs=gr.Textbox(lines=2, placeholder="Write your qeustion here....."),
    outputs="text",
    title="AI assistant(Qwen + LangChain)",
    description="Ask the llm any questions and It will be  give you the answer"
)

# Display
demo.launch()
```
